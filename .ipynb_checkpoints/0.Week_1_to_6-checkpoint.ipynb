```r
co2.values = as.numeric(co2)
co2.values
```

```r
time(co2)
```

```r
co2.times = as.numeric(time(co2))
```

Dependent variable: Time
Target : CO2

```r
SSxx = sum((co2.times - mean(co2.times))**2)  # Sum of squares of X 
SSxy = sum((co2.times - mean(co2.times))*(co2.values - mean(co2.values))) 
(slope = SSxy / SSxx)
(intercept = mean(co2.values) - slope*mean(co2.times))
```

```r
#Linear model
co2.linear.model = lm(co2 ~ time(co2) )
plot(co2, main='Atmospheric CO2 Concentration with Fitted Line')
abline(co2.linear.model )
```

```r
# Compute residuals 
co2.fitted.values = slope*co2.times + intercept
co2.residuals = co2.values - co2.fitted.values
```

```r
# You might be a little lazier and would like R to do this for you. We interrogate the linear model with a command like
( co2.residuals = resid( co2.linear.model ) )
```

# Sleep data

A very famous data set was discussed in William Gossetts (hes better known as Student) classic paper The Probable Error of a Mean (Student, 1908). Gossett describes an experiment in which two sleeping aides are administered during the course of a clinical trial to a group of 10 research subjects. The average number of hours of sleep gained (called by the variable name
extra) under the use of each of the two drugs (indicated by the variable name group) are recorded. When R starts it has a number of packages already loaded. This data set should be available to you just by typing, at the R command console, 

```r
sleep
```

```r
plot(extra~group, data=sleep, main = "Extra Sleep in Gossett Data by Group")
```

If you are tired of typing the dollar sign to access a variable, you can attach the data frame as

```r
attach(sleep)
```

Now we can access the variables with a little less typing. Disaggregating the extra sleep data into variables specifying those who took drug 1 and those who took drug 2 

```r
extra.1=extra[group==1]
extra.2=extra[group==2]
```

```r
# Run a t-test
t.test(extra.1, extra.2, paired=TRUE, alternative="two.sided")
```

Lets think about all this in another way. Since the data are paired (two different responses by same subject to each drug), it is natural to think of them in an (, ) fashion and plot our data in a scatter plot. What do we observe? 

```r
plot(extra.2~extra.1, xlab='extra sleep with drug 1', ylab='extra sleep with drug 2' ,
     main='Extra Sleep Drug 2 against Extra Sleep Drug 1')
sleep.linear.model = lm(extra.2 ~ extra.1 )
abline(sleep.linear.model)
```

We have changed our focus from (a) deciding whether the group averages differ to (b) thinking about whether we might say something about a persons response to the second drug given their
response to the first drug. Aside from the one pesky data point (subject 9) it looks like we might be able to do a decent job.
So, in this context we might think about performing a regression of 2 response on 1 response.

```r
summary(sleep.linear.model)
```

## Simulating random walk

```r
x = NULL
x[1] = 0
for(i in 2:1000){
    x[i]=x[i-1]+rnorm(1)
}

random_walk = ts(x)
plot(random_walk, main='A random walk', ylab='', xlab='Days', lwd=2, col='blue')
```

```r
acf(random_walk)
```

The first difference of random walk should be stationary

```r
plot(diff(random_walk))
```

```r
acf(diff(random_walk))
```

This is stationary

## Moving average process 

Intuition using a stock price example:

$X_t$ is a stock price of some company. Each daily announcement of the company to the press is modeled as a noise. Announcements affect the stock price. If there is an announcement today, well, it will have an effect on the stock price and it is possible that effect of the announcement might last a few days.

Let's say, each announcements affects the price two days, then we can think of stock price as a linear combination of the noises until two days back. 

It is possible that yesterday's noise, yesterday's announcement might have less effect. So, we have a weight on it. This is called the moving average model of order two. Now, you can think of MA(q) model, where $X_t$ is a linear combination of these noises few days back and I have $\theta_1$, $\theta_2$ .. $\theta_q$. These are the weights of the noises from yesterday and so forth. And Zis here are basically independent, identically distributed random variables and they are modeled as a normal random variable with some mean and standard deviation. 

```r
# Generate noise
noise=rnorm(10000)

# Introduce a variable
ma_2=NULL

# Loop for generating MA(2) process

for(i in 3:10000){
    ma_2[i]=noise[i]+0.7*noise[i-1]+0.2*noise[i-2]
}

# Shift data to left by 2 units
moving_average_process=ma_2[3:10000]

# Put time series structure on a vanilla data
moving_average_process=ts(moving_average_process)

# Partition output graphics as a multi frame of 2 rows and 1 column
par(mfrow=c(2,1))

# plot the process and plot its ACF
plot(moving_average_process, main='A moving average process of order 2', ylab=' ', col='blue')
acf(moving_average_process, main='Correlogram of a moving average process of order 2')
```

```r
# Simulating MA(4) process.
# X_t= Z_t+0.2 Z_(t-1)+0.3 Z_(t-2)+ 0.4 Z_(t-3)

set.seed(2^10)
z=NULL
z=rnorm(1000)
data=NULL
for(i in 4:1000){
  data[i-3]=z[i]+0.2*z[i-1]+0.3*z[i-2]+0.4*z[i-3]
  }
data=ts(data)

# find acf below

(acf(data))
```
